# Data Cleaning 


Steps:
*   Convert datetimes
*   Convert installers not in the top 5 to other
* Convert funder not in the top 5 to other 




```
#Drop Nulls and update df_values 
df_values['date_recorded'] = pd.to_datetime(df_values['date_recorded'])
df_values['construction_year'] = pd.to_datetime(df_values['construction_year'])

df_values[['date_recorded', 'construction_year']].dtypes
```

```
#replace all installers who arent in the top 5 with other 
```

```
'DWE'                   
'Government'             
'RWE'                   
'Commu'                  
'DANIDA'   
'Other'             
```

```
def installer_cleaner(row):
#Goes thru each row and replaces anything not in the top 5 installers with other'

  if row['installer'] == 'DWE':
    return 'DWE'
  elif row['installer'] == 'Government':
    return 'Danida'
  elif row['installer'] == 'RWE':
    return 'RWE'
  elif row['installer'] == 'Commu':
    return 'Commu'
  elif row['installer'] == 'DANIDA':
    return 'DANIDA'
  else:
    return 'other'

#run formula 
df_values['installer'] = df_values.apply(lambda row: installer_cleaner(row), axis=1) 
```

```
df_values['installer'].unique()
```

```
funders = df_values['funder'].value_counts().rename_axis('funder').reset_index(name='counts')
funders['funder'].iloc[0:5]
```

```
Government Of Tanzania
Danida
Hesawa
Rwssp
World Bank
```

```
def funder_cleaner(row):  
  #Same as before but for funders

    if row['funder']=='Government Of Tanzania':
        return 'Government of Tanzania'
    elif row['funder']=='Danida':
        return 'Danida'
    elif row['funder']=='Hesawa':
        return 'Hesawa'
    elif row['funder']=='Rwssp':
        return 'Rwssp'
    elif row['funder']=='World Bank':
        return 'World_Bank'    
    else:
        return 'other'

#run formulas 
df_values['funder'] = df_values.apply(lambda row: funder_cleaner(row), axis=1)
```

```
df_values.isna().sum()
```

```
df_values['public_meeting'] = df_values['public_meeting'].fillna('Unknown')
```

```
df_values['scheme_management'].value_counts()
```

```
VWC                 36793
WUG                  5206
Water authority      3153
WUA                  2883
Water Board          2748
```

```
def scheme_management_cleaner(row):  
  #Same as before but for funders

    if row['funder']=='VWC':
        return 'VWC'
    elif row['funder']=='WUG':
        return 'WUG'
    elif row['funder']=='Water authority':
        return 'Water authority'
    elif row['funder']=='WUA':
        return 'WUA'
    elif row['funder']=='Water Board':
        return 'Water Board'    
    else:
        return 'other'

#run formulas 
df_values['scheme_management'] = df_values.apply(lambda row: scheme_management_cleaner(row), axis=1)
```

```
df_values['permit'] = df_values['permit'].fillna('Unknown')
```

# Random Forest 

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```
x = train.drop(columns=['construction_year'], axis=1)

y = df_labels['status_group']

x
```

```
# train test split

x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                     train_size = 0.25, random_state=123)
```

```
tree_clf = DecisionTreeClassifier(criterion='gini', max_depth=5)
tree_clf.fit(x_train, y_train)
```

```
# Make predictions for test data
y_pred = tree_clf.predict(x_test)  
y_pred
```

```
from sklearn.metrics import accuracy_score, roc_curve, auc

# Calculate accuracy 
acc = accuracy_score(y_test,y_pred) * 100
print('Accuracy is :{0}'.format(acc))
```
