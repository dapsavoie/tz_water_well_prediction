# Build Decision Tree Classifier 

```
dummies_train = pd.read_csv(r'/content/drive/My Drive/tz_water_well_data/train_dummmies.csv')

dummies_test = pd.read_csv(r'/content/drive/My Drive/tz_water_well_data/test_dummies.csv')
```

```
x = dummies_train

y= df_labels['status_group']
```

```
from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(x, y, 
                                                                    test_size = 0.25, random_state=123)
```

```
from sklearn.tree import DecisionTreeClassifier

# Instantiate and fit a DecisionTreeClassifier
tree_clf2 = DecisionTreeClassifier(random_state=10, criterion='entropy')  
tree_clf2.fit(data_train, target_train)
```

```
y_pred= tree_clf2.predict(data_test) #does not shuffle index of data 
y_pred
```

```
test_preds = tree_clf2.predict(dummies_test)
test_preds
```

```
data_test['pred'] = y_pred

data_test.columns
```

```
from sklearn.metrics import accuracy_score
# Calculate accuracy 
acc = accuracy_score(target_test,y_pred)
print('Accuracy is :{0}'.format(acc))
```
